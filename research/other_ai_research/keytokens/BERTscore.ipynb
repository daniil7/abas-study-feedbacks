```python
from bert_score import BERTScorer
import pandas as pd
```

```python
scorer = BERTScorer(lang="ru")
```

```python
feedbacks_tokens_predictions = pd.read_csv("drive/MyDrive/ОБЩЕЕ/Отзывус/AI/key_tokens/data/feedbacks_tokens_predictions_KeyBERT_DeepPavlov.csv").dropna()
feedbacks_tokens_predictions
```

```python
P, R, F1 = scorer.score(list(feedbacks_tokens_predictions.TFIDF), list(feedbacks_tokens_predictions.Feedbacks))
f"TFIDF. Precision: {P.mean():.3f} Recall: {R.mean():.3f} F1 score: {F1.mean():.3f}"
```

```python
P, R, F1 = scorer.score(list(feedbacks_tokens_predictions.RAKE), list(feedbacks_tokens_predictions.Feedbacks))
f"RAKE. Precision: {P.mean():.3f} Recall: {R.mean():.3f} F1 score: {F1.mean():.3f}"
```

```python
P, R, F1 = scorer.score(list(feedbacks_tokens_predictions.YAKE), list(feedbacks_tokens_predictions.Feedbacks))
f"YAKE. Precision: {P.mean():.3f} Recall: {R.mean():.3f} F1 score: {F1.mean():.3f}"
```

```python
P, R, F1 = scorer.score(list(feedbacks_tokens_predictions.TextRank), list(feedbacks_tokens_predictions.Feedbacks))
f"TextRank. Precision: {P.mean():.3f} Recall: {R.mean():.3f} F1 score: {F1.mean():.3f}"
```
